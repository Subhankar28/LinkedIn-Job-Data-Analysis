# Exploratory Data Analysis (E.D.A.)

**Job Openings As Per State**

```python
plt.figure(figsize=(13, 6), dpi=150)
chart1 = sns.countplot(df["Location"], palette='mako')
plt.title("Job Openings by State", size=25)
plt.ylabel("Number of job openings", size=20)
plt.xlabel("State", size=20)
plt.xticks(rotation=80)
chart1.bar_label(chart1.containers[0], size=12)
```

**Jobs Available In Each Industry**

```python
plt.figure(figsize = (13,6), dpi = 150)
chart2 = sns.countplot(df['Industry'], palette= 'twilight')
plt.title('Job openings By Industry', size = 22)
plt.ylabel('Number Of Job Openings', size=20)
plt.xlabel('Industry', size=20)
plt.xticks(rotation=30, fontsize=12)
chart2.bar_label(chart2.containers[0], size=12)
```

**Jobs Available In Company As Per Employee Count**

```python
plt.figure(figsize=(13, 6), dpi=150)
chart3 = sns.countplot(df["Name"], palette='Blues_r')
plt.ylabel("Number of job openings", size=20)
plt.xlabel("Company", size=20)
plt.xticks(rotation=70)
chart3.bar_label(chart3.containers[3], size=13)
```

**Job Openings At Each Level**

```python
plt.figure(figsize=(13, 6), dpi=150)
chart4 = sns.countplot(df["Level"], palette='ocean')
plt.title("Job Openings at different levels", size=25)
plt.ylabel("Number of job openings", size=20)
plt.xlabel("Level", size=20)
plt.xticks(rotation=70)
chart4.bar_label(chart4.containers[0], size=13)
```

**Job Openings For Different Involvement Nature**

```python
plt.figure(figsize=(13, 6), dpi=200)
chart5 = sns.countplot(df["Involvement"], palette='ocean_r')
plt.title("Job Openings at different Involvement Nature", size=25)
plt.ylabel("Number of job openings", size=20)
plt.xlabel("Involvement_Nature", size=20)
plt.xticks(rotation=70)
chart5.bar_label(chart5.containers[0], size=13)
```

**Checking Relation Between Linkedin Follower And Employee Count**

```python
plt.figure(figsize=(20, 10), dpi=100)
chart6 = sns.scatterplot(data = df, x="LinkedIn_Followers", y="Employee_count", hue="Name", s=400, palette='Dark2')
plt.title("Relation between LinkedIn followers & employee count for each company", size=25)
plt.ylabel("Employee Count", size=22)
plt.xlabel("Followers", size=22)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
chart6.bar_label(chart6.containers[0], size=13)
```

**Number Of Applications Received By Companies.**

```python
group_data = df[["Name", 'LinkedIn_Followers', "Total_applicants"]].groupby(["Name", 'LinkedIn_Followers'])
sum_applicants = group_data.sum()
Applications = sum_applicants.reset_index()
Applications
```

```python
plt.figure(figsize=(20,10),dpi=100)
chart7 = sns.barplot(data=Applications,x="Name",y="Total_applicants", palette='ocean')
plt.ylabel("Application count",size=20)
plt.xlabel("Company",size=20)
plt.xticks(rotation=60,size=15)
plt.yticks(size=15)
chart7.bar_label(chart7.containers[0],size=15)
```

**Jobs that require Specific particular skill**

```python
df_skills = df.drop(["Name", "Designation", "Location", "Total_applicants", "LinkedIn_Followers", "Involvement", "Involvement", "Level", "Employee_count", "Industry"], axis=1)

skill_dic = {}

for i in df_skills.columns:
    skill_dic[i] = df[i].sum()

data = list(skill_dic.items())

skill_df = pd.DataFrame(data, columns=['Skill', 'Number of jobs requiring the skill'])
skill_df
```

```python
plt.figure(figsize=(20,10),dpi=100)
chart8 = sns.barplot(x=skill_df["Skill"],y=skill_df["Number of jobs requiring the skill"],palette="icefire")
plt.title("Number of jobs that require a particular skill",size=25)
plt.ylabel("Number of jobs",size=22)
plt.xlabel("Skill",size=22)
plt.xticks(fontsize=18,rotation=90)
plt.yticks(fontsize=18)
chart8.bar_label(chart8.containers[0],size=18)
```
